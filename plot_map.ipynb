```python
import pandas as pd
import geopandas
import folium
from folium import plugins
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"E:\Desenvolvimento\Teste\acumulado_MT.xlsx", usecols=['VL_LAT', 'VL_LONG','ID_CONDI'])
```

```python
df = df[:50]
```

```python
df = df.loc[df['ID_CONDI'] != 0]
df = df.loc[df['VL_LAT'] != '0']
```

```python
df.dropna(inplace=True)
```

```python
status = {
    1:'Alive',
    2:'Sulk',
    3:'Died'
}

df['ID_CONDI'] = df['ID_CONDI'].apply(lambda x: status[x])
```

```python
df['VL_LAT'] = df['VL_LAT'].apply(lambda x: x.replace(',','.'))
df['VL_LONG'] = df['VL_LONG'].apply(lambda x: x.replace(',','.'))
```

```python
geometry = geopandas.points_from_xy(df['VL_LONG'], df['VL_LAT'])
geo_df = geopandas.GeoDataFrame(df[['ID_CONDI','VL_LAT','VL_LONG']], geometry=geometry)
```

```python
world = geopandas.read_file(r'E:\Desenvolvimento\Teste\brazil_polygon\CEMbrMUNa20.shp')
```

```python
fig, ax = plt.subplots(figsize=(12,9))
world.plot(ax=ax, alpha=0.4, color='grey')
geo_df.plot(column='ID_CONDI', ax=ax, legend=True)
plt.title('Ninhos')
```

```python
map = folium.Map(tiles='Cartodb dark_matter', zoom_start = 10, location=[df['VL_LAT'][0], df['VL_LONG'][0]])
```

```python
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry ]
# geo_df['geometry'].to_file('./result/cordenadas.json', driver="GeoJSON")
geo_df
```

```python
# Iterate through list and add a marker for each volcano, color-coded by its type.
for index, coordinates in enumerate(geo_df_list):
    #assign a color marker for the type of volcano, Strato being the most common
    if geo_df.ID_CONDI[index] == "Alive":
        type_color = "green"
    elif geo_df.ID_CONDI[index] == "Sulk":
        type_color = "orange"
    elif geo_df.ID_CONDI[index] == "Died":
        type_color = "red"
    else:
        type_color = "cyan"


    plugins.HeatMap(geo_df_list, min_opacity=0.8).add_to(map)
map
```
